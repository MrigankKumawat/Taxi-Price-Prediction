### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

### Loading the Data

```python
df = pd.read_csv("Dataset/taxi_trip_pricing.csv")
```

### Handling Null and Duplicate Values

```python
df.isnull().sum()
```

```python
## For Numeric Columns
df['Trip_Distance_km'] = df['Trip_Distance_km'].fillna(df['Trip_Distance_km'].median())
df['Trip_Price'] = df['Trip_Price'].fillna(df['Trip_Price'].median())
df['Trip_Duration_Minutes'] = df['Trip_Duration_Minutes'].fillna(df['Trip_Duration_Minutes'].median())
```

```python
## For Categorical Columns
for col in ['Time_of_Day', 'Day_of_Week', 'Passenger_Count', 'Traffic_Conditions', 'Weather']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# Machine Learning

```python
# Importing Libraries for Machine Learning
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import cross_val_score
```

### Stratified Shuffle Split

```python
df['distance_cat'] = pd.cut(df['Trip_Distance_km'], bins = [0, 25, 50, 75, 100, np.inf], labels = [1, 2, 3, 4, 5])
```

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_set, test_set in split.split(df, df['distance_cat']):
    strat_train_set = df.loc[train_set]
    strat_test_set = df.loc[test_set]
```

### Copying the Train Set

```python
taxi_train = strat_train_set.copy()
```

### Separating Features and Labels

```python
taxi_features_train = taxi_train.drop(["Trip_Price", "distance_cat"], axis = 1)
taxi_label_train = taxi_train['Trip_Price']
```

### Separating Numerical and Categorical Values

```python
num_attrbs_train = ['Trip_Distance_km', 'Passenger_Count', 'Trip_Duration_Minutes']
cat_attrbs_train = ['Time_of_Day', 'Day_of_Week', 'Traffic_Conditions', 'Weather']
```

## Let's Start Making Pipelines

### Numeric and Categorical Pipelines

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy = "median")),
    ("scaler", StandardScaler()),
])
cat_pipeline = Pipeline([
    ("onehot", OneHotEncoder(handle_unknown = "ignore")),
])
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attrbs_train),
    ("cat", cat_pipeline, cat_attrbs_train),
])
```

### Transforming the Data

```python
taxi_prepared_train = full_pipeline.fit_transform(taxi_features_train)
```

### Training the Model

#### Linear Regression Model

```python
lin_model = LinearRegression()
lin_model.fit(taxi_prepared_train, taxi_label_train)
lin_predict = lin_model.predict(taxi_prepared_train)
lin_CV = cross_val_score(lin_model, taxi_prepared_train, taxi_label_train, scoring = "neg_root_mean_squared_error", cv = 10)
lin_R2 = r2_score(taxi_label_train, lin_predict)
lin_MAE = mean_absolute_error(taxi_label_train, lin_predict)
lin_RMSE = root_mean_squared_error(taxi_label_train, lin_predict)

print("Using Linear Regression Model for Training Data -:")
print(pd.Series(lin_CV).describe())
print(f"\nR2 Score:{lin_R2:.2f}")
print(f"MAE Score:{lin_MAE:.2f}")
print(f"RMSE Score:{lin_RMSE:.2f}")
```

#### Random Forest Regressor Model

```python
forest_model = RandomForestRegressor(
    n_estimators = 200,
    max_depth = 12,
    min_samples_split = 8,
    min_samples_leaf = 4,
    max_features = "sqrt",
    random_state = 42
)
forest_model.fit(taxi_prepared_train, taxi_label_train)
forest_predict = forest_model.predict(taxi_prepared_train)
forest_CV = cross_val_score(forest_model, taxi_prepared_train, taxi_label_train, scoring = "neg_root_mean_squared_error", cv = 10)
forest_R2 = r2_score(taxi_label_train, forest_predict)
forest_MAE = mean_absolute_error(taxi_label_train, forest_predict)
forest_RMSE = root_mean_squared_error(taxi_label_train, forest_predict)

print("Using Random Forest Regressor Model for Training Data -:")
print(pd.Series(forest_CV).describe())
print(f"\nR2 Score:{forest_R2:.2f}")
print(f"MAE Score:{forest_MAE:.2f}")
print(f"RMSE Score:{forest_RMSE:.2f}")
```

#### Gradient Boosting Regressor Model

```python
grad_model = GradientBoostingRegressor(
    n_estimators = 300,
    learning_rate = 0.04,
    max_depth = 3,
    min_samples_split = 9,
    min_samples_leaf = 4,
    subsample = 0.8,
    max_features = "sqrt"
)
grad_model.fit(taxi_prepared_train, taxi_label_train)
grad_predict = grad_model.predict(taxi_prepared_train)
grad_CV = cross_val_score(grad_model, taxi_prepared_train, taxi_label_train, scoring = "neg_root_mean_squared_error", cv = 10)
grad_R2 = r2_score(taxi_label_train, grad_predict)
grad_MAE = mean_absolute_error(taxi_label_train, grad_predict)
grad_RMSE = root_mean_squared_error(taxi_label_train, grad_predict)

print("Using Gradient Boosting Regressor Model for Training Data -:")
print(pd.Series(grad_CV).describe())
print(f"\nR2 Score:{grad_R2:.2f}")
print(f"MAE Score:{grad_MAE:.2f}")
print(f"RMSE Score:{grad_RMSE:.2f}")
```

## Testing the Model - For the Best Training Model which is Gradient Boosting
